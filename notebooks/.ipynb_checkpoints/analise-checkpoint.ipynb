# Análise de músicas do Spotify

## Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dados/data.csv')
```

```python
df.info()
```

```python
df.describe().T
```

```python
plt.figure(figsize = (10,6))
df['year'].plot(kind = 'hist')
plt.show()
```

```python
for x in list(df):
    try:
        plt.figure(figsize = (8,5))
        df[x].plot(kind = 'hist')
        plt.title(x)
        plt.show()
    except:
        pass
```
